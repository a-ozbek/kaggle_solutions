```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_test = pd.read_json('./data/test.json')
df_sub = pd.read_csv('./submissions/sub5-6-7-8-10-11-12-18-19-25.csv')
```

```python
df = pd.merge(df_test, df_sub, on='id')
df
```

```python
df['is_iceberg'].hist()
```

```python
# Visualize images
iceberg_cond = (df['is_iceberg'] > 0.9) & (df['is_iceberg'] < 1.0)
not_iceberg_cond = (df['is_iceberg'] > 0.0) & (df['is_iceberg'] < 0.1)
df_ = df[iceberg_cond | not_iceberg_cond].sample(n=20).sort_values(by='is_iceberg')
print len(df[iceberg_cond | not_iceberg_cond])
for im_band1, im_band2, angle in zip(df_['band_1'], df_['band_2'], df_['is_iceberg']):    
    #
    im_band1 = np.array(im_band1).reshape(75, 75)
    im_band2 = np.array(im_band2).reshape(75, 75)
    # Plot
    fig = plt.figure(figsize=(10, 6))
    plt.subplot(1,3,1); plt.imshow(im_band1); plt.title('Band1, ' + str(angle))
    plt.subplot(1,3,2); plt.imshow(im_band2); plt.title('Band2, ' + str(angle))
    plt.subplot(1,3,3); plt.imshow((im_band1 + im_band2) / 2.0);
```

```python
df['is_m_generated'] = df['inc_angle'].apply(lambda x: len(str(x)) > 7)
```

```python
#
fig = plt.figure(figsize=(15, 6))
plt.subplot(1,2,1); 
abs(df[df['is_m_generated'] == True]['is_iceberg'] - 0.5).hist();
plt.title('Machine Generated')
plt.subplot(1,2,2);
abs(df[df['is_m_generated'] == False]['is_iceberg'] - 0.5).hist()
plt.title('Not Machine Generated')
```

```python
t = 0.075
iceberg_cond = (df['is_iceberg'] >= 1.0 - t) & (df['is_iceberg'] <= 1.0)
not_iceberg_cond = (df['is_iceberg'] >= 0.0) & (df['is_iceberg'] <= t)
df[iceberg_cond | not_iceberg_cond]['is_m_generated'].value_counts()
```

```python
# Visualize machine generated images
# Iceberg
df_ = df[(iceberg_cond) & (df['is_m_generated'] == True)].sample(n=20).sort_values(by='is_iceberg')
for im_band1, im_band2, angle in zip(df_['band_1'], df_['band_2'], df_['is_iceberg']):    
    #
    im_band1 = np.array(im_band1).reshape(75, 75)
    im_band2 = np.array(im_band2).reshape(75, 75)
    # Plot
    fig = plt.figure(figsize=(10, 6))
    plt.subplot(1,3,1); plt.imshow(im_band1); plt.title('Band1, ' + str(angle))
    plt.subplot(1,3,2); plt.imshow(im_band2); plt.title('Band2, ' + str(angle))
    plt.subplot(1,3,3); plt.imshow((im_band1 + im_band2) / 2.0);
```

```python
# Not - Iceberg
df_ = df[(not_iceberg_cond) & (df['is_m_generated'] == True)].sample(n=20).sort_values(by='is_iceberg')
for im_band1, im_band2, angle in zip(df_['band_1'], df_['band_2'], df_['is_iceberg']):    
    #
    im_band1 = np.array(im_band1).reshape(75, 75)
    im_band2 = np.array(im_band2).reshape(75, 75)
    # Plot
    fig = plt.figure(figsize=(10, 6))
    plt.subplot(1,3,1); plt.imshow(im_band1); plt.title('Band1, ' + str(angle))
    plt.subplot(1,3,2); plt.imshow(im_band2); plt.title('Band2, ' + str(angle))
    plt.subplot(1,3,3); plt.imshow((im_band1 + im_band2) / 2.0);
```
